```python
# import the relevant libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the data
df_grade = pd.read_csv('df_grade.csv')
df_grade_ma = pd.read_csv('df_grade_ma.csv')
```

```python
# Review the mean, median, variance, standard deviation, minimum and maximum score of the data
print(f'The number of students in BSCSIU4 is {df_grade_ma.shape[0]}.\nThe mean of Lab 1 in BSCSIU4 is {df_grade_ma["Lab 1"].mean()}.\nThe variance of Lab 1 in BSCSIU4 is {df_grade_ma["Lab 1"].var()}.\nThe standard deviation of Lab 1 in BSCSIU4 is {df_grade_ma["Lab 1"].std()}.\nThe minimum of Lab 1 in BSCSIU4 is {df_grade_ma["Lab 1"].min()}.\nThe maximum of Lab 1 in BSCSIU4 is {df_grade_ma["Lab 1"].max()}.\nThe median of Lab 1 in BSCSIU4 is {df_grade_ma["Lab 1"].median()}.\n')
print(f'The number of students of the whole class is {df_grade.shape[0]}.\nThe mean of Lab 1 of the whole class is {df_grade["Lab 1"].mean()}.\nThe variance of Lab 1 of the whole class is {df_grade["Lab 1"].var()}.\nThe standard deviation of Lab 1 of the whole class is {df_grade["Lab 1"].std()}.\nThe minimum of Lab 1 of the whole class is {df_grade["Lab 1"].min()}.\nThe maximum of Lab 1 of the whole class is {df_grade["Lab 1"].max()}.\nThe median of Lab 1 of the whole class is {df_grade["Lab 1"].median()}.\n')
```

```python
# Visualize the data with density histograms and compare via boxplots.
# Create a figure with one subplot
fig, ax = plt.subplots(figsize=(10, 5))

# Plot the density histograms for df_grade and df_grade_ma on the same subplot
ax.hist([df_grade['Lab 1'], df_grade_ma['Lab 1']], bins=10, density=True, label=['Whole Class', 'BSCSIU4'])
ax.set_xlabel('Lab 1 score')
ax.set_ylabel('Density')
ax.set_title('Lab 1 score distribution for the whole class and BSCSIU4')
ax.legend()

# Show the plot
plt.show()

# Create a figure with one subplot
fig, ax = plt.subplots(figsize=(10, 5))

# Plot the boxplots for df_grade and df_grade_ma on the same subplot
ax.boxplot([df_grade['Lab 1'], df_grade_ma['Lab 1']])
ax.set_xticklabels(['Whole Class', 'BSCSIU4'])
ax.set_title('Lab 1 score for the whole class and BSCSIU4')

# Show the plot
plt.show()
```
